```python
#@markdown <font color = 'brown'> **STEP 3️⃣ : Split words <font color = 'sky blue'> 단어로 쪼개기** 🅰️ ✔️ 🅱️

stopwords = set(STOPWORDS)
spltxt = text.split()
```

```python

#@markdown <font color = 'green'> **Step 4️⃣ : Create a wordcloud <font color = 'sky blue'> 워드클라우드 만들기** 🚾 ☁️ ⚡

import matplotlib.pyplot as plt
from wordcloud import WordCloud
import random

# 1. Define your specific color palette
def custom_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    colors = ["#e74c3c", "#2ecc71", "#3498db", "#9b59b6", "#e67e22"] # Red, Green, Blue, Purple, Orange
    return random.choice(colors)

# 2. Configure the WordCloud
wordcloud = WordCloud(
    stopwords=stopwords,
    relative_scaling=0.1,        # Lower values make the size difference less extreme
    random_state=3,
    max_words=1000,              # Reduced from 2000 to prevent "crowding"
    background_color='white',
    width=1600,
    height=800,
    prefer_horizontal=0.85,      # Makes it much easier for the audience to read
    font_step=1,
    min_font_size=10
).generate(' '.join(spltxt))

# 3. Apply the custom colors
wordcloud.recolor(color_func=custom_color_func)

# 4. Display and Save
plt.figure(figsize=(15, 7.5), facecolor='white')
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()

wordcloud.to_file('vietnam_recruitment_cloud.png')
wordcloud.to_file('vietnam_recruitment_frequency_cloud.png')
```
